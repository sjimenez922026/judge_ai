## Análisis de Frontera Bayesiana

La **frontera bayesiana** (Bayesian Frontier Analysis) nos permite:
* Identificar las unidades más eficientes (profesionales o IAs)
* Medir la distancia de cada unidad respecto a la frontera de eficiencia
* Comparar el desempeño relativo considerando incertidumbre

### Enfoque:
Usaremos un modelo bayesiano que considera:
1. **Input**: Tiempo de creación (recurso consumido)
2. **Outputs**: Scores de calidad (métricas de evaluación)
3. **Eficiencia**: Ratio output/input ajustado por incertidumbre

```python
class BayesianFrontierAnalysis:
    """
    Análisis de Frontera Bayesiana para comparar eficiencia
    entre profesionales e IAs en la creación de logotipos.
    """
    
    def __init__(self, evaluations: List[EvaluationResult]):
        self.evaluations = evaluations
        self.df = self._prepare_data()
    
    def _prepare_data(self) -> pd.DataFrame:
        """Prepara los datos para el análisis"""
        data = []
        for eval_result in self.evaluations:
            row = {
                'nombre': eval_result.submission.nombre_creador,
                'tipo': eval_result.submission.tipo_creador,
                'tiempo': eval_result.submission.tiempo_creacion,
                'score_total': eval_result.metrics.score_total(),
                'creatividad': eval_result.metrics.creatividad,
                'relevancia': eval_result.metrics.relevancia,
                'impacto_visual': eval_result.metrics.impacto_visual
            }
            data.append(row)
        return pd.DataFrame(data)
    
    def calculate_efficiency_scores(self) -> pd.DataFrame:
        """
        Calcula scores de eficiencia: output (calidad) / input (tiempo)
        """
        if self.df.empty:
            return pd.DataFrame()
        
        df_efficiency = self.df.copy()
        
        # Eficiencia simple: score / tiempo
        df_efficiency['efficiency_simple'] = df_efficiency['score_total'] / df_efficiency['tiempo']
        
        # Eficiencia normalizada (0-1)
        max_eff = df_efficiency['efficiency_simple'].max()
        df_efficiency['efficiency_normalized'] = df_efficiency['efficiency_simple'] / max_eff
        
        # Clasificar unidades
        threshold = df_efficiency['efficiency_normalized'].quantile(0.75)
        df_efficiency['categoria'] = df_efficiency['efficiency_normalized'].apply(
            lambda x: 'Frontera' if x >= threshold else 'Por debajo de frontera'
        )
        
        return df_efficiency.sort_values('efficiency_normalized', ascending=False)
    
    def visualizar_frontera(self):
        """Visualiza la frontera de eficiencia"""
        if self.df.empty:
            print("⚠️ No hay datos para visualizar")
            return
        
        df_eff = self.calculate_efficiency_scores()
        
        fig, axes = plt.subplots(1, 2, figsize=(16, 6))
        
        # Gráfico 1: Frontera de eficiencia (Tiempo vs Score)
        ax1 = axes[0]
        for tipo in df_eff['tipo'].unique():
            df_tipo = df_eff[df_eff['tipo'] == tipo]
            color = '#FF6B6B' if tipo == 'ia' else '#4ECDC4'
            marker = 'o' if tipo == 'ia' else 's'
            ax1.scatter(df_tipo['tiempo'], df_tipo['score_total'], 
                       s=200, alpha=0.6, color=color, marker=marker,
                       label=tipo.capitalize(), edgecolors='black', linewidth=1.5)
        
        # Añadir nombres
        for idx, row in df_eff.iterrows():
            ax1.annotate(row['nombre'], 
                        (row['tiempo'], row['score_total']),
                        xytext=(5, 5), textcoords='offset points',
                        fontsize=8, alpha=0.7)
        
        ax1.set_xlabel('Tiempo de Creación (minutos)', fontsize=12, fontweight='bold')
        ax1.set_ylabel('Score Total de Calidad', fontsize=12, fontweight='bold')
        ax1.set_title('Frontera de Eficiencia: Tiempo vs Calidad', fontsize=14, fontweight='bold')
        ax1.legend(fontsize=10)
        ax1.grid(True, alpha=0.3)
        
        # Gráfico 2: Eficiencia normalizada
        ax2 = axes[1]
        colors = ['#FF6B6B' if t == 'ia' else '#4ECDC4' for t in df_eff['tipo']]
        bars = ax2.barh(df_eff['nombre'], df_eff['efficiency_normalized'], color=colors, alpha=0.7)
        ax2.axvline(x=0.75, color='red', linestyle='--', linewidth=2, label='Umbral Frontera (Q3)')
        ax2.set_xlabel('Eficiencia Normalizada', fontsize=12, fontweight='bold')
        ax2.set_title('Ranking de Eficiencia (Score/Tiempo)', fontsize=14, fontweight='bold')
        ax2.legend()
        ax2.grid(True, alpha=0.3, axis='x')
        
        plt.tight_layout()
        display(plt.show())
        
        return df_eff
    
    def generar_reporte(self) -> Dict:
        """Genera un reporte completo del análisis"""
        df_eff = self.calculate_efficiency_scores()
        
        reporte = {
            'total_evaluaciones': len(df_eff),
            'por_tipo': df_eff['tipo'].value_counts().to_dict(),
            'eficiencia_promedio': {
                'general': df_eff['efficiency_normalized'].mean(),
                'por_tipo': df_eff.groupby('tipo')['efficiency_normalized'].mean().to_dict()
            },
            'en_frontera': df_eff[df_eff['categoria'] == 'Frontera']['nombre'].tolist(),
            'mejor_ia': df_eff[df_eff['tipo'] == 'ia'].iloc[0]['nombre'] if len(df_eff[df_eff['tipo'] == 'ia']) > 0 else None,
            'mejor_profesional': df_eff[df_eff['tipo'] == 'profesional'].iloc[0]['nombre'] if len(df_eff[df_eff['tipo'] == 'profesional']) > 0 else None
        }
        
        return reporte

print("✓ Clase BayesianFrontierAnalysis definida")
```

```python
@dataclass
class LogoSubmission:
    """Representa una propuesta de logotipo"""
    id: str
    tipo_creador: str  # 'profesional' o 'ia'
    nombre_creador: str  # Nombre del profesional o modelo IA
    descripcion: str  # Descripción del concepto
    prompt_usado: Optional[str]  # Solo para IAs
    tiempo_creacion: float  # Tiempo en minutos
    timestamp: datetime
    imagen_path: Optional[str]  # Ruta a la imagen del logotipo
    
    def to_dict(self) -> Dict:
        return {
            'id': self.id,
            'tipo_creador': self.tipo_creador,
            'nombre_creador': self.nombre_creador,
            'descripcion': self.descripcion,
            'prompt_usado': self.prompt_usado,
            'tiempo_creacion': self.tiempo_creacion,
            'timestamp': self.timestamp.isoformat(),
            'imagen_path': self.imagen_path
        }

@dataclass
class EvaluationResult:
    """Resultado completo de una evaluación"""
    submission: LogoSubmission
    metrics: EvaluationMetrics
    juez_id: str  # Identificador del juez/evaluador
    comentarios: Optional[str]
    
    def to_dict(self) -> Dict:
        return {
            'submission': self.submission.to_dict(),
            'metrics': self.metrics.to_dict(),
            'juez_id': self.juez_id,
            'comentarios': self.comentarios
        }

print("✓ Estructuras de datos definidas")
```

```python
class LogoJudgeSystem:
    """Sistema juez para evaluar logotipos"""
    
    def __init__(self):
        self.evaluations: List[EvaluationResult] = []
        self.ai_models = ['ChatGPT', 'Gemini', 'DeepSeek', 'Qwen', 'Claude']
        
    def add_evaluation(self, evaluation: EvaluationResult):
        """Agrega una evaluación al sistema"""
        self.evaluations.append(evaluation)
        print(f"✓ Evaluación agregada: {evaluation.submission.nombre_creador} - Score: {evaluation.metrics.score_total():.2f}")
    
    def get_results_dataframe(self) -> pd.DataFrame:
        """Convierte las evaluaciones a un DataFrame de pandas"""
        if not self.evaluations:
            return pd.DataFrame()
        
        data = []
        for eval_result in self.evaluations:
            row = {
                'id': eval_result.submission.id,
                'tipo_creador': eval_result.submission.tipo_creador,
                'nombre_creador': eval_result.submission.nombre_creador,
                'tiempo_creacion': eval_result.submission.tiempo_creacion,
                **eval_result.metrics.to_dict(),
                'juez_id': eval_result.juez_id
            }
            data.append(row)
        
        return pd.DataFrame(data)
    
    def get_summary_statistics(self) -> pd.DataFrame:
        """Genera estadísticas resumidas por tipo de creador"""
        df = self.get_results_dataframe()
        if df.empty:
            return pd.DataFrame()
        
        summary = df.groupby(['tipo_creador', 'nombre_creador']).agg({
            'score_total': ['mean', 'std', 'min', 'max', 'count'],
            'creatividad': 'mean',
            'relevancia': 'mean',
            'impacto_visual': 'mean',
            'tiempo_creacion': 'mean'
        }).round(2)
        
        return summary
    
    def save_results(self, filepath: str):
        """Guarda los resultados en formato JSON"""
        data = [eval_result.to_dict() for eval_result in self.evaluations]
        with open(filepath, 'w', encoding='utf-8') as f:
            json.dump(data, f, indent=2, ensure_ascii=False)
        print(f"✓ Resultados guardados en: {filepath}")

# Inicializar el sistema juez
judge_system = LogoJudgeSystem()
print(f"✓ Sistema Juez inicializado")
print(f"  Modelos IA a evaluar: {', '.join(judge_system.ai_models)}")
```

- **Métricas**: Creatividad, relevancia, impacto visual, alineación con público objetivo
- **Frontera Bayesiana**: Análisis de eficiencia relativa

```python
# Configuración e imports necesarios
import pandas as pd
import numpy as np
from dataclasses import dataclass
from typing import List, Dict, Optional
from datetime import datetime
import json

print("✓ Librerías importadas correctamente")
```

```python
# Definición de métricas de evaluación
@dataclass
class EvaluationMetrics:
    """Métricas para evaluar diseños de logotipos"""
    creatividad: float  # 0-10: Originalidad y pensamiento innovador
    relevancia: float  # 0-10: Alineación con brief y marca
    impacto_visual: float  # 0-10: Atractivo estético y memorabilidad
    coherencia_marca: float  # 0-10: Consistencia con identidad de marca
    publico_objetivo: float  # 0-10: Adecuación al target
    aplicabilidad: float  # 0-10: Versatilidad en diferentes medios
    
    def score_total(self) -> float:
        """Calcula el score ponderado total"""
        return (
            self.creatividad * 0.20 +
            self.relevancia * 0.20 +
            self.impacto_visual * 0.20 +
            self.coherencia_marca * 0.15 +
            self.publico_objetivo * 0.15 +
            self.aplicabilidad * 0.10
        )
    
    def to_dict(self) -> Dict:
        return {
            'creatividad': self.creatividad,
            'relevancia': self.relevancia,
            'impacto_visual': self.impacto_visual,
            'coherencia_marca': self.coherencia_marca,
            'publico_objetivo': self.publico_objetivo,
            'aplicabilidad': self.aplicabilidad,
            'score_total': self.score_total()
        }

print("✓ Métricas de evaluación definidas")
```
